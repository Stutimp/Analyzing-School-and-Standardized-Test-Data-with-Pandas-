```python
# Calculate the total school budget and per capita spending per school
#calculating the total school budget for each school
per_school_budget = school_data.groupby("school_name")["budget"].sum().reset_index()
#per_school_budget
```

```python
#After renaming ther column
per_school_budget = per_school_budget.rename(columns={"budget":"Total School Budget"})
per_school_budget
```

```python
school_budget = pd.merge(per_school_counts,per_school_budget, how="outer", on=["school_name", "school_name"])
school_budget
```

```python
# Calculate the average test scores per school
#for math score using groupby and using reset_index
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean().reset_index()

#Renaming the column for clarity
per_school_math = per_school_math.rename(columns={"math_score": "Average Math Score"})
#merging the average math score with school_budget dataframe
school_budget = school_budget.merge(per_school_math,on = "school_name")
school_budget
```

```python
#now for reading score
# using groupby and using reset_index
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean().reset_index()
#renaming the colunmn for clarity
per_school_reading =per_school_reading.rename(columns={"reading_score":"Average Reading Score"})
#merging the average reading score with school_budget dataframe
school_budget= school_budget.merge(per_school_reading, on="school_name")
school_budget
```
